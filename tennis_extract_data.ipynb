# Odds 

```python
import pandas as pd
from pathlib import Path

base_path = Path('D:/Trains/tennis_project/data/raw_data')
subfolder = 'raw_odds_parquet'  

all_data = []

for date_folder in sorted(base_path.iterdir()):
    match_folder = date_folder / subfolder
    if not match_folder.exists():
        continue

    for file in match_folder.glob("*.parquet"):
        try:
            df = pd.read_parquet(file, engine="fastparquet")
            all_data.append(df)
        except Exception as e:
            print(f"An error occured: {file.name}  {e}")
        print(f'{file.name} is concated, date: {date_folder.name}')

final_odd_df = pd.concat(all_data, ignore_index=True)
```
